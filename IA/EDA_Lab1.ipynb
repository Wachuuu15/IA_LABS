```python
df.domain_in_title.value_counts(normalize=True).plot(kind="bar")

plt.title("Dominio URL en el título")
plt.xlabel("Presencia (Positiva/Negativa)")
plt.ylabel("Porcentaje")
plt.xticks(rotation=0)
```

```python
df.domain_with_copyright.value_counts(normalize=True).plot(kind="bar")

plt.title("Dominio URL dentro del Copyright")
plt.xlabel("Presencia (Negativa/Positiva)")
plt.ylabel("Porcentaje")
plt.xticks(rotation=0)
```

```python
df.google_index.value_counts(normalize=True).plot(kind="bar")

plt.title("Indexación a Google")
plt.xlabel("Presencia (Positiva/Negativa)")
plt.ylabel("Porcentaje")
plt.xticks(rotation=0)
```

```python
df.status.value_counts().plot.barh(color = "red")

plt.title("Estado final de la URL")
plt.xlabel("Frecuencia")
plt.ylabel("Estado final")
plt.xticks(rotation=0)
```
